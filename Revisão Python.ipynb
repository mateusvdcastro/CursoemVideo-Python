# **Funções, arquivos e pacotes**

### **Funções - o que é uma função?**

Funções, em linguagens de programação, são trechos de código que podemos reutilizar em qualquer parte do código, facilitando a legibilidade e a escalabilidade. Sem funções e métodos nossos códigos seriam todos uma coisa só, e seria difícil interpretar o código como um todo, ver aonde se repetem rotinas, códigos e entre outros.

As funções também são poderosas mesmo sem haver repetição no código, pois nos permite além de dar nomes para porções específicas do código nós permite editar a funcionalidade de uma parte do código em uma parte separada no arquivo. Ou seja, não precisamos procurar aonde a função está sendo utilizada. Atualizar na função já atualiza no programa inteiro

### **Vantagens das funções**
1. Escalabilidade
2. Refatoração
3. Reutilização
4. Facilidade em editar funcionalidades no código inteiro
5. Visualização, possibilidade de dar nomes

Usamos funções no código inteiro. Todo tipo de rotina que fizermos durante o código, inclusive dentro de funções, deve ser chamado por uma função menor e mais especializada. 

Um código com poucas funções é um código mal fatorado. E um código com muitas funções, se usadas da maneira correta, é um código legível e bem fatorado.

Dica: funções devem ter nomes de verbos no imperativo, seja em inglês ou português. Isso é pois a função deve ser lida como um comando, ou seja, algo que o programador está mandando o programa fazer. Por isso, é comum evitar nomes como "somar" e usar "soma" ou "some" por exemplo.
 


```
def some(a, b):
  # liberdade para escrever qualquer código aqui
  return a + b
```

```
some(1, 2)
```

```
1 + 2
```

### **Parâmetros opcionais**

É possível definir parâmetros opcionais em Python. \\

Mas qual é o uso disso? Voltando a função de soma de a e b, suponha que queremos deixar o usuário escolher printar de uma forma mais organizada o retorno da função ou apenas retornar a soma. Como fazer isso em python sem obrigar o usuário a mandar True ou False toda vez?

### **O que representa se um trecho de código pertence ou não a uma função**

Perceba que o código possui uma identação, aqui no jupyter essa indentação é de 2 espaços, nos outros editores o padrão é de 4 espaços. \\
Cada indentação, representa a entrada de um contexto. Quando você indenta para inserir um trecho de código dentro um `if`, é similar a inserir um trecho de código dentro de uma função. Então essa estrutura pode ser comparada ao trecho de código dentro de condicionais e loops.

```
def some_print(a, b, prprint=False):
  soma = a + b

  if prprint:
    print(f'A soma de {a} e {b} é igual a {soma}')
    return
  
  return soma
```

```
# outro jeito de escrever a função
def some_print(a, b, prprint=False):
  soma = a + b

  if not prprint:
    return soma
  
  print(f'A soma de {a} e {b} é igual a {soma}')  

# Perceba! Uma função sem retorno retorna None, e não necessariamente precisamos
# ter argumentos.
  
```

```
# Teste 1
some_print(1, 2)
```

```
# Teste 2
some_print(1, 2, prprint=True)
```

```
# Teste 3
some_print(1, 2, True)
```

```
# erro
some_print(1)
```

- **`some_print`**: nome da variável da nova função \\
- **`a, b`**: parâmetros obrigatórios da função \\

*Lembrem-se: um parâmetro é opcional quando o mesmo tem um valor atribuido na assinatura da função. Neste caso, prprint recebeu o valor False* \\

Repare no corpo da função: \\
1. Temos a variável soma, que armazena a soma para ser reutilizada depois \\
2. Temos uma checagem para ver qual é a decisão do programador, e ai reagimos baseado no valor de prprint. Se printamos e não retornamos nada, ou apenas retornamos. \\

Repare: podemos não retornar nada se quisermos. Afinal, em python, não precisamos declarar o retorno de uma função se não quisermos e no caso esse função pode retornar um int ou None. Nesse caso, o return foi utilizado para sair da função mais cedo.

```
var = 0

def func():
  print(f'Sou um erro! (var) {var}')
  var = 1
  print(f'Dentro de func(): var = {var}')

func()
print(f'Fora de func(): var = {var}')
```

### **Podemos tirar os erros acima usando a keyword global**

Ao usar global dentro de algum escopo, você define que o escopo que a váriavel (por exemplo) x irá utilizar para leitura e escrita não é o seu escopo local, mas sim o escopo global.

```
var = 0

def func():
  global var
  print(f'Após global: var = {var}')
  var = 1
  print(f'Dentro de func(): var = {var}')

func()
print(f'Fora de func(): var = {var}')
```

# **Abrindo e fechando arquivos**

### **`open()`**
A função `open()` recebe o caminho relativo do arquivo para abrir como argumento. \\
Ou seja, o caminho a partir de onde o terminal que o python está sendo executado está. \\
Além do caminho para o arquivo, caso o programador queira fazer algo além de simplesmente ler ele deve modificar o parâmetro opcional "mode" (que é o segundo parâmetro da função). \\
Existem os seguintes valores para mode.
1. **`'r'`** (padrão): abre um arquivo para leitura
2. **`'w'`**: Abre um arquivo para escrita, se já existir, apaga ele e começa a escrever do 0. Se não existir, o cria
4. **`'a'`**: Abre um arquivo para escrita no final. Se não existir o cria
5. **`'b'`**: Abre um arquivo binário ao invés de um arquivo de texto
7. **`'+'`**: Combinado com os modos anteriores, atualiza para o modo de atualização, isto é, leitura e escrita

Existem outros modos, mas não abordaremos eles nesse curso. Para saber mais, abra a documentação ou apenas help(open) no python.

A função retorna um objeto de arquivo.

### **`arquivo.close()`**

Ao terminar de fazer as operações com o arquivo selecionado, é necessário pedir ao sistema operacional para fechar o arquivo. Para isso, apenas use a variável retornada e chame o método `.close()` do arquivo.

### **Jeito alternativo, e melhor de se lidar com abertura de arquivos**

Ao invés de fechar um arquivo manualmente, existe um jeito em python de se fazer isso automaticamente, através da palavra chave with. Será explicada em uma celula de código.

```
# celula para baixar os arquivos que iremos manipular
!wget -O lorem.txt "drive.google.com/uc?id=1lZp-nEm-FyWQarU6K4l9mKP3-PQicZZj&export=download"
!wget -O numeros.txt "drive.google.com/uc?id=1X8Yrmdf5nKiLkVO0-zHhiQjcxEA5qGqA&export=download"
```

```
arq = open('./lorem.txt') # abre apenas para a leitura
arq.close()

arq = open('./numeros.txt', 'r+') # abre para a leitura e escrita
arq.close()

arq = open('./novo.txt', 'w') # abre apenas para a escrita
arq.close()

arq = open('./numeros.txt', 'a') # abre para escrita no final
arq.close()
```

### **Leitura de arquivos**

### **`arquivo.read()`**

Le o arquivo inteiro e armazena em uma única string

```
with open('./lorem.txt') as arq:
  lorem_string = arq.read()

with open('./numeros.txt') as arq:
  numeros_string = arq.read()

print(lorem_string)
print(numeros_string)
```

### **`arquivo.readline()`**
Lê linha por linha do arquivo, avançando pra próxima linha

```
with open('./lorem.txt') as arq:
  linha = arq.readline()

  while linha:
    print(linha.strip())
    linha = arq.readline()
```

### **`arquivo.readlines()`**
Lê o arquivo inteiro, mas armazena um vetor com todas as linhas

```
with open('./lorem.txt') as arq:
  linhas = arq.readlines()

print(linhas)
print([linha.strip() for linha in linhas])
```

### **Exemplo: lendo uma tabela de números separados por espaço**

```
with open('./numeros.txt') as arq:
  array = [[int(x) for x in line.split()] for line in arq]
  print(array)
```

### **Escrita em arquivos**

### **`arquivo.write(string)`**

Escreve no arquivo a string enviada na função, na posição atual, que pode ser controlada através do seek

### **`arquivo.seek(offset: numero)`**

Muda a posição atual do arquivo através de um offset de bytes, que é relativo aos valores da posição. \\
Valores: \\
`0` -> Começo do arquivo \\
`1` -> Posição atual \\
`2` -> Fim do arquivo \\

### **Exemplo: escrevendo em um arquivo (truncando)**

```
with open('./lorem.txt') as arq:
  lorem_txt = arq.read()

with open('./novo.txt', 'w+') as arq:
  arq.write(lorem_txt)
  
  arq.seek(0)
  print(arq.read())
```

### **Exemplo: adicionando conteúdo no final do arquivo**

```
with open('./numeros.txt') as arq:
  numeros_string = arq.read()

with open('./novo.txt', 'a+') as arq:
  arq.write(numeros_string)

  arq.seek(0)
  print(arq.read())
```

# **NumPy**

### **O que é o NumPy?**

O NumPy é uma biblioteca feita para a linguagem Python de alta performance que adiciona suporte para trabalho com matrizes e vetores em várias dimensões.

### **O que é um array?**

Um Array do numpy é a implementação da lista do python em alta performance gerenciado pelo numpy. Ou seja, é uma sequencia de valores indexados por uma tupla de dimensão n formada por inteiros não negativos.

### **Como criar um array a partir de uma lista**

Para isso, precisamos usar a função `np.array()`, veja:


```
import numpy as np

lista = [1, 2, 3, 4, 5, 6]
ndarray = np.array(lista)

print(ndarray)
```

### **Outras maneiras de criar um array**

1. **`np.zeros`**: preenche um array com zeros no formato Z
2. **`np.ones`**: preenche um array com uns no formato Z
3. **`np.arange`**: preenche um array com valores de X a Y, a cada passo Z
4. **`np.linspace`**: preenche um array com Z valores de X a Y igualmente espaçados
5. **`np.random.randint`**: preenche um array com valores aleatórios de X a Y no formato Z


```
zeros = np.zeros((3, 3))
print(zeros)
```

```
ones = np.ones((3, 3))
print(ones)
```

```
arange = np.arange(0, 10, 3)
print(arange)
```

```
linspace = np.linspace(0, 50, 10)
print(linspace)
```

```
randint = np.random.randint(-8, 8, (3, 3))
print(randint)
```

### **Algumas propriedades dos arrays**
1. **`ndim`**: retorna a dimensão do array
2. **`shape`**: retorna o formato do array

### **Algumas operações trigonométricas no NumPy**

1. **`np.pi`**: retorna o valor de pi
2. **`np.cos`**: calcula o cosseno de um número ou todos os cossenos de um array
3. **`np.sin`**: calcula o seno de um número ou todos os senos de um array
4. **`np.tan`**: calcula a tangente de um número ou todas as tangentes de um array

```
print(f'Valor de pi: {np.pi}\n')

print(f'Valor do cosseno de 2 * pi: {np.cos(2 * np.pi)}')
print(f'Valor do seno de 2 * pi: {np.sin(2 * np.pi)}')
print(f'Valor da tangente de 2 * pi: {np.tan(2 * np.pi)}')
```

```
wave = np.linspace(0, 2 * np.pi, 5)
print(f'Valores de 0, 90, 180, 270 e 360 graus em radianos: \n{wave}')

print(f'\nSenos desses valores: \n{np.sin(wave)}')
print(f'\nCossenos desses valores: \n{np.cos(wave)}')
print(f'\nTangentes desses valores: \n{np.tan(wave)}')
```

### **Outras operações com arrays do NumPy**

1. **`np.sum`**: realiza a soma de todos os valores de um array
2. **`np.median`**: encontra a mediana dentro de um array
3. **`np.mean`**: encontra a média dentro de um array
4. **`np.std`**: encontra o desvio padrão dado um array
5. **`np.sort`**: ordena um array
6. **`np.abs`**: transforma um número negativo em positivo (não acontece nada se o número já for positivo) ou transforma todos os números negativos em positivos dentro de um array
7. **`np.random.shuffle`**: reordena as posições do array de forma aleatória


```
seq = np.arange(1, 11)

print(f'Sequencia de 1 a 10: {seq}')
print(f'Soma dos numeros da sequencia: {np.sum(seq)}')
print(f'Mediana da sequência: {np.median(seq)}')
print(f'Média da sequência: {np.mean(seq)}')
print(f'Desvio padrão da sequência: {np.std(seq)}')
```

```
np.random.shuffle(seq)
print(f'Sequência após o shuffle \n{seq}')

seq = np.sort(seq)
print(f'\nSequência após o sort \n{seq}')
```

```
randarr = np.random.randint(-10, 11, 10)
print(f'Sequência aleatória de -10 a 10 \n{randarr}')

print(f'\nApós o np.abs(): \n{np.abs(randarr)}')
```

### **`Slice`**

É possível fazer slice de arrays assim como é possível fazer slice de listas em Python. Ou seja, `array[start:stop[:step]]`.
- **`start`**: o indice que começa o slice
- **`stop`**: o indice que termina
- **`step`**: o "passo", ou seja, de quantos em quantos indices é desejado exibir

```
seq = np.arange(-100, 101)
print(seq)
```

### **Exemplo: indíces que repesentam os valores pares de 0 a 100**

```
print(seq[100:201:2])
```

### **Exemplo indíces que representam os valores de 50 a 100**

```
print(seq[150:201])
```

### **Slice condicional**

Além do Slice tradicional, o numpy oferece o slice condicional. Ou seja, é enviado uma condição para realizar um slice no array, **baseado no valor
e não no índice** \\
Para isso, é necessario fazer um conjunto de condições com operadores não ambíguos (`==`, `!=`, `&` para `and`, `|` para `or`).

### **Exemplo: valor positivo é igual a 100**

```
print(seq[abs(seq) == 100])
```

### **Exemplo: valores diferentes de 0**

```
print(seq[seq != 0])
```

### **Exemplo: valores maiores que 50**

```
print(seq[seq > 50])
```

### **Exemplo: valores no intervalo (-inf, -50] ou [50, inf)**

```
print(seq[(seq <= -50) | (seq >= 50)])
```

### **Exemplo: valores no intervalo [-50, 50]**

```
print(seq[(seq >= -50) & (seq <= 50)])
```

### **Reproduzindo o operador de adição do NumPy**


```
array1 = np.array([1, 2, 3, 4])
array2 = np.array([5, 6, 7, 8])

array3 = array1 + array2
array4 = np.concatenate([array1, array2])

list1 = list(array1)
list2 = list(array2)

from operator import add
list3 = list(map(add, list1, list2))
list4 = list1 + list2

print(f'(np.ndarray) array1 + array2: {array3}',
      f'\n(np.ndarray) np.concatenate(array1, array2): {array4}')
print(f'\n(list) Equivalente da soma do NumPy usando map: {list3}',
      f'\n(list) list1 + list2: {list4}')
```

# **Comparativo de performance**

Comparação inspirada em: https://stackoverflow.com/questions/18713321/element-wise-addition-of-2-lists \\

```
import timeit
from operator import add
import numpy as np

list1 = [4, 5, 6]*10**5
list2 = [1, 2, 3]*10**5

array1 = np.array(list1)
array2 = np.array(list2)

def run_map_add():
  list(map(add, list1, list2))

def run_arr_add():
  array1 + array2

print('Tempo sem NumPy (1 execução):', timeit.timeit(run_map_add, number=1))
print('Tempo com NumPy (1 execução):', timeit.timeit(run_arr_add, number=1))
print('Tempo sem NumPy (200 execuções):', timeit.timeit(run_map_add, number=200))
print('Tempo com NumPy (200 execuções):', timeit.timeit(run_arr_add, number=200))
print('Tempo com NumPy (10000 execuções):', timeit.timeit(run_arr_add, number=10000))
```

# **Introdução ao Pandas**

O [Pandas](https://pandas.pydata.org) é uma biblioteca de código aberto amplamente utilizada por cientistas e analistas de dados, isso se dá pelo fato de que ela facilita a manipulação de dados tabulados (como um Excel ou tabela de banco de dados). Alguns benefícios que a biblioteca fornece são:

* Estruturas de dados e operações similares à uma planilha comum (Series e Dataframe);
* Funções prontas para processamento e "limpeza" de dados;
* Funções prontas para manipulação de diferentes tipos de arquivos (csv, xls, xlsx, json, etc)

### **Hello World!**

A fim de termos o primeiro contato com as estruturas de dados básicas, vamos criar as séries e o dataframe exibidos na imagem ilustrativa e realizar algumas operações básicas.

Obs.: Nesse primeiro exercício também iremos realizar a importação da biblioteca.

```
import pandas as pd # A partir de agora podemos referenciar a biblioteca do Pandas com o "pd"
```

Criando as duas séries com os nomes "Apples" e "Orange"

```
apples_serie = pd.Series(data=[3,2,0,1], name='Apples')
oranges_serie = pd.Series(data=[0,3,7,2], name='Oranges')

print(apples_serie)
print(oranges_serie)
```

É importante notar que as séries não são simples lista.

```
type(apples_serie)
```

Os operadores básicos (+, -, * e /) podem ser utilizados para operar duas séries, gerando uma nova série. Nesse caso, cada valor da primeira série é operado com o valor correspondente (mesmo índice) na segunda série.

```
apples_serie + oranges_serie
# apples_serie - oranges_serie
# apples_serie * oranges_serie
# apples_serie / oranges_serie
```

Para criarmos o Dataframe podemos passar as duas séries em formato de dicionário, onde a **chave** será o nome da série (coluna) e o **valor** será a série em si.

```
fruits_df = pd.DataFrame(data={
                        'Apples': apples_serie,
                        'Oranges': oranges_serie
                        })
fruits_df
```

Também podemos adicionar séries (colunas) à um dataframe de maneira bastante simples.

```
total_fruits = apples_serie + oranges_serie
fruits_df['Total Fruits'] = total_fruits
fruits_df
```

# **Explorando um dataset**

A fim de se familiarizar um pouco mais com a estrutura do DataFrame, vamos carregar e trabalhar com um dataset mais interessante.

Todas as funções disponíveis no Pandas pode ser encontrada [aqui](https://pandas.pydata.org/docs/reference/frame.html)

Uma grande facilidade oferecida pelo Pandas é o carregamento de dados de diferentes fontes. Para exemplificar isso, além de termos algo mais interessante para trabalhar, vamos carregar uma base diretamente de uma [planilha](https://docs.google.com/spreadsheets/d/1Bq5IxppilddbsQ0aVAToRkVdeQ0i9KlEsWPLq-fCaQg/edit?usp=sharing) do Google Drive (Retirada do [Kaggle](https://www.kaggle.com/gregorut/videogamesales)).

```
# Não se preocupe com esse trecho de código
# Ele básicamente vai cuidar da autorização de acesso e 
# de ler os dados da planilha para a varável "rows"
# ===========================================================
from google.colab import auth
import gspread
from oauth2client.client import GoogleCredentials
auth.authenticate_user()

GAMES_SALES_DATASET_URL = 'https://docs.google.com/spreadsheets/d/1Bq5IxppilddbsQ0aVAToRkVdeQ0i9KlEsWPLq-fCaQg/edit?usp=sharing'

gc = gspread.authorize(GoogleCredentials.get_application_default())
worksheet = gc.open_by_url(GAMES_SALES_DATASET_URL).sheet1
rows = worksheet.get_all_values()
# ===========================================================

# Vamos dar uma olhada nos dados que estão em "rows"
# Note que é uma simples lista de listas, onde o primeiro item é uma lista com
# os dados de cabeçalho e os demais itens são as linhas de dados
print(type(rows))
print(rows[0])
print(rows[1])
print(rows[2])
```

Uma vez que os dados da planilha foi carregada para a variável "rows", vamos utiliza-la para criar um DataFrame do Pandas

```
# Podemos utilizar a função from_records() para definir as linhas e cabecalhos
g_sales_df = pd.DataFrame.from_records(columns=rows[0], data=rows[1:])
```

A partir desse ponto, a variável g_sales_df possui um DataFrame do Pandas com os dados obtidos da planilha referenciada. Vamos investigar um pouco esses dados utilizando algumas funções disponíveis.

```
# A função head() pode ser utilizada visualizar apenas as primeiras linhas do
# DataFrame

g_sales_df.head() # Por default, são exibidas 5 linhas, mas podemos definir quantas queremos
```

```
# A função describe() apresenta um resumo sobre os dados de cada coluna no DataFrame
# As informações apresentadas variam com o tipo de dados da coluna.

g_sales_df.describe()
```

```
# Vamos dar uma olhada nos tipos de dados de cada coluna
g_sales_df.dtypes
```

```
# Visando evitar esse tipo genérico (object), vamos converter as colunas
# para tipos de dados mais interessantes
g_sales_df['Rank'] = g_sales_df['Rank'].astype('int')
g_sales_df['Name'] = g_sales_df['Name'].astype('category')
g_sales_df['Platform'] = g_sales_df['Platform'].astype('category')
g_sales_df['Year'] = g_sales_df['Year'].astype('category')
g_sales_df['Genre'] = g_sales_df['Genre'].astype('category')
g_sales_df['Publisher'] = g_sales_df['Publisher'].astype('category')
g_sales_df['NA_Sales'] = g_sales_df['NA_Sales'].astype('float')
g_sales_df['EU_Sales'] = g_sales_df['EU_Sales'].astype('float')
g_sales_df['JP_Sales'] = g_sales_df['JP_Sales'].astype('float')
g_sales_df['Other_Sales'] = g_sales_df['Other_Sales'].astype('float')
g_sales_df['Global_Sales'] = g_sales_df['Global_Sales'].astype('float')
g_sales_df.dtypes
```

# **Agrupando dados**

Quando trabalhamos com os dados categóricos em uma planilha ou em um banco de dados, a realização de agrupamentos e sumariazações é bastante comum e útil. O Pandas oferece algumas alternativas para esse tipo de operação.

### **Group-By**

Operação similar à operaçã de mesmo nome utilizada na linguagem SQL para bancos de dados relacionais.

```
# Obtendo a soma de vendas para cada plataforma
g_sales_df.groupby(by=['Platform']).sum()
```

```
# Obtendo a média de vendas para cada genero
g_sales_df.groupby(by=['Genre']).mean()
```

```
# Obtendo a soma de vendas para cada genero em cada ano
g_sales_df.groupby(by=['Genre', 'Year']).sum()
```

```
# Extra - Podemos usar a função 'loc' para acessar os valores dos anos de um genero específico
vendas_por_gen_ano = g_sales_df.groupby(by=['Genre', 'Year']).sum()
vendas_por_gen_ano.loc['Racing'].head(10) # Apenas os 10 primeiros
```

### **Pivot Tables**

Formatação bastante utilizada em planilhas.

```
# Criando uma Pivot table que irá mostras a soma das vendas na North America para cada Ano, Plataforma e Genero
# Obs.: O parâmetro fill_value, indica para a função substituir por 0 qualquer valor faltante
g_sales_df.pivot_table(values='NA_Sales', index=['Year', 'Platform'], columns=['Genre'], aggfunc='sum', fill_value=0)
```

### **Componentes dos Gráficos**


![chart_components.png](https://matplotlib.org/_images/anatomy.png)

```
# fig: É a "Figure", basicamente é imagem/figura como um todo e contém os demais elementos.
# ax: É o "Axes", onde a área de dados é desenhada.
# Importante: "Axes" != "Axis".
# Um "Axes" pode ter 2 ou 3 eixos ("Axis")
# os "Axis" cuida dos limites e ticks

fig, ax = plt.subplots()
ax.plot(range(10), label='linear')
ax.set_xlabel('x label')
ax.set_xlim(left=0, right=15)
ax.set_ylabel('y label')
ax.set_ylim(bottom=0, top=20)
ax.set_title("Gráfico Simples")
ax.legend() # Apenas faz a legenda ser exibida, os dados vem do label da curva
ax.grid() # Draw grid lines
# fig.show() Nesse caso, ao invés de plt.show() podemos chamar o show() da "Figure"
```

# **Gráficos de Barras**

[API](https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.axes.Axes.bar.html#matplotlib.axes.Axes.bar)

```
# Vamos plotar as vendas de 3 produtos nos quarters de 2019
quarters = ['Q1/2019', 'Q2/2019', 'Q3/2019', 'Q4/2019'] # Eixo x
vendas_pcs = [10, 13, 11, 16]
vendas_laptops = [24, 23, 27, 32]
vendas_video_games = [8, 6, 7, 10]
largura = 0.5

fig, ax = plt.subplots()
ax.bar(x=quarters, height=vendas_pcs, width=largura, label='PCs')

# Para "empilhar" uma barra, marcamos seu inicio no final da anterior com o parâmetro "bottom"
ax.bar(x=quarters, height=vendas_laptops, width=largura, bottom=vendas_pcs, label='Laptops')

# Para "empilhar" a terceirca barra, precisamos definir onde as outras últimas barras terminam
# A função zip() retorna tuplas com os valores das listas de um mesmo índice
#print(list(zip(vendas_pcs, vendas_laptops)))
temp_bottom = [a+b for a,b in zip(vendas_pcs, vendas_laptops)]
ax.bar(x=quarters, height=vendas_video_games, width=largura, bottom=temp_bottom, label='Video Games')

ax.set_ylabel('Total de Vendas')
ax.set_xlabel('Quarter')
ax.set_title('Vendas por Quarter e Produto')
ax.legend()
```

# **Multiplos Gráficos**

### **Multiplas curvas em 1 gráfico**

```
# Dados para plots
quarters = ['Q1/2019', 'Q2/2019', 'Q3/2019', 'Q4/2019']
vendas_pcs = [10, 13, 11, 16]
vendas_laptops = [24, 23, 27, 32]
vendas_video_games = [8, 6, 7, 10]
bottom_2 = [a+b for a,b in zip(vendas_pcs, vendas_laptops)]
lucro = [3715.76, 3910.23, 4407.53, 6131.59]
largura = 0.5

# Gerando barras
fig, ax_qtd = plt.subplots()
ax_qtd.bar(x=quarters, height=vendas_pcs, width=largura, label='PCs')
ax_qtd.bar(x=quarters, height=vendas_laptops, width=largura, bottom=vendas_pcs, label='Laptops')
ax_qtd.bar(x=quarters, height=vendas_video_games, width=largura, bottom=bottom_2, label='Video Games')

# Gerando linhas
# Gerando novo eixo y com o mesmo x
ax_lucro = ax_qtd.twinx()
ax_lucro.plot(lucro, '--om', label='Lucro', linewidth=4, markersize=8)

# Adicionando Informações no gráfico
ax_qtd.set_xlabel('Quarter')
ax_qtd.set_ylabel('Venda Total')

ax_lucro.set_ylabel('Lucro Total')
ax_lucro.set_ylim(0, 7000)

ax_qtd.set_title('Vendas por Quarter e Produto')
ax_qtd.legend()
ax_lucro.legend(loc='upper center')
fig.show()
```

### **Multiplos gráficos em 1 figura**

```
# Dados para plots
quarters = ['Q1/2019', 'Q2/2019', 'Q3/2019', 'Q4/2019']
vendas_pcs = [10, 13, 11, 16]
vendas_laptops = [24, 23, 27, 32]
vendas_video_games = [8, 6, 7, 10]
products = ['PCs', 'Laptops', 'Video Games']
vendas = [10, 24, 8]
bottom_2 = [a+b for a,b in zip(vendas_pcs, vendas_laptops)]
lucro = [3715.76, 3910.23, 4407.53, 6131.59]
largura = 0.5

# Gerando barras
fig, axs = plt.subplots(2, 2) # 2 linha e 2 colunas

# Quando temos linhas > 1 e colunas > 1, acessamos como uma matriz [x][y]
axs[0][0].bar(x=quarters, height=vendas_pcs, width=largura, label='PCs')
axs[0][0].bar(x=quarters, height=vendas_laptops, width=largura, bottom=vendas_pcs, label='Laptops')
axs[0][0].bar(x=quarters, height=vendas_video_games, width=largura, bottom=bottom_2, label='Video Games')

axs[0][0].set_xlabel('Quarter')
axs[0][0].set_ylabel('Venda Total')
axs[0][0].set_title('Vendas por Quarter e Produto')
axs[0][0].legend()

# Gerando linha
# Gerando novo eixo y com o mesmo x
axs[0][1].plot(quarters, lucro, '--om', label='Lucro', linewidth=4, markersize=8)
axs[0][1].set_ylabel('Lucro Total')
axs[0][1].set_ylim(0, 7000)
axs[0][1].set_title('Lucro')
axs[0][1].legend()

# Gerando pizza
axs[1][0].pie(vendas, labels=products, autopct='%.2f%%', shadow=True)
axs[1][0].set_title('Proporção de venda de produtos')

# Aumentando o tamanho da figura para não distorcer os gráficos
fig.set_figwidth(12)
fig.set_figheight(6)

fig.tight_layout() # Adiciona um padding entre os gráficos automaticamente

fig.show()
```

# **Pandas + Matplotlib**

O Pandas utiliza o Matplotlib para facilitar a visualização de seus DataFrames, a seguir vamos investigar um pouco essa integração.

[API](https://pandas.pydata.org/docs/reference/frame.html#plotting)

```
# Carregando os dados
import pandas as pd
from google.colab import auth
import gspread
from oauth2client.client import GoogleCredentials
auth.authenticate_user()

GAMES_SALES_DATASET_URL = 'https://docs.google.com/spreadsheets/d/1Bq5IxppilddbsQ0aVAToRkVdeQ0i9KlEsWPLq-fCaQg/edit?usp=sharing'

gc = gspread.authorize(GoogleCredentials.get_application_default())
worksheet = gc.open_by_url(GAMES_SALES_DATASET_URL).sheet1
rows = worksheet.get_all_values()

g_sales_df = pd.DataFrame.from_records(columns=rows[0], data=rows[1:])
g_sales_df['Name'] = g_sales_df['Name'].astype('category')
g_sales_df['Platform'] = g_sales_df['Platform'].astype('category')
g_sales_df['Year'] = g_sales_df['Year'].astype('category')
g_sales_df['Genre'] = g_sales_df['Genre'].astype('category')
g_sales_df['Publisher'] = g_sales_df['Publisher'].astype('category')
g_sales_df['NA_Sales'] = g_sales_df['NA_Sales'].astype('float')
g_sales_df['EU_Sales'] = g_sales_df['EU_Sales'].astype('float')
g_sales_df['JP_Sales'] = g_sales_df['JP_Sales'].astype('float')
g_sales_df['Other_Sales'] = g_sales_df['Other_Sales'].astype('float')
g_sales_df['Global_Sales'] = g_sales_df['Global_Sales'].astype('float')
g_sales_df.head()
```

A função plot() do DataFrame faz um plot básico de todas as séries numéricas.

```
g_sales_df.plot()
```

Vamos trabalhar um pouco os dados para melhorar a visualização.

```
# Deixando apenas linhas com o valor dos anos
g_sales_df = g_sales_df[g_sales_df['Year'] != 'N/A']
g_sales_df['Year'] = g_sales_df['Year'].astype('int')

# Ordernando pelo ano
g_sales_df = g_sales_df.sort_values(by='Year')

# Tornando a coluna 'Year' como indice
indexed_df = g_sales_df.set_index('Year')

# Removendo coluna Rank
indexed_df = indexed_df.drop(columns='Rank')

indexed_df.head()
```

```
# Soma de vendas por ano
indexed_df.groupby(by=['Year']).sum().plot()
```

```
# Soma de vendas por genero
indexed_df.groupby(by=['Genre']).sum().plot.area()
```

```
# Média de vendas por plataforma
indexed_df.groupby(by=['Platform']).mean().plot.bar(figsize=(20, 5))
```

```
# Relação de vendas entre Plataforma e Genero
indexed_df.plot.scatter(x='Platform', y='Genre', figsize=(20, 5), s='Global_Sales')
```

```
# Relação de vendas entre North America e Europa
indexed_df.plot.scatter(x='NA_Sales', y='EU_Sales', figsize=(6, 6), s='Global_Sales', xlim=(0, 40), ylim=(0,40))
```
